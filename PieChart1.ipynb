```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

slices = [ 14, 36, 14, 36]

labels = ['14% OTHERS', '36% Lack of Personal Interest', '14% High Cost of Education', '36% Illness or Disability']

colors = ['#cf9f9f', '#dcb6b6', '#f6d4c9', '#66545e']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'white'}, colors=colors)

plt.tight_layout()
plt.grid(True)
plt.show()
```
