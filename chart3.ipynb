```python
import matplotlib.pyplot as plt

labels = ['89% of the worlds student population']
share = [89] # sales in million units
# plt.pie(x=share, labels=labels, autopct='%.2f%%')

# explode = (0.1, 0, 0, 0, 0)  
plt.style.use('ggplot')
plt.pie(x=share, labels=labels, 
         startangle=90)

   

marks = [10, 45, 45]


my_circle = plt.Circle((0, 0), .75, color='white')

# Give color names
plt.pie(marks,
        colors=['#43BFC7','#008B8B', '#20B2AA'], startangle=90)

p = plt.gcf()
p.gca().add_artist(my_circle)

# Show the graph
plt.show()
```
